```python
import pandas as pd
import matplotlib.pyplot as plt
import settings
```

# Connect to the database

offers_df = pd.read_sql_table('offers','sqlite:///offers.db', index_col='id')

# Offers choice

offers_bmw_seria5 = offers_df[offers_df['model'] == 'Seria 5']

offers_bmw_seria5.head()

# Group by series

bmw_grouped_offers = offers_bmw_seria5.groupby('production_year')

# Show basic statistics 

bmw_grouped_offers.course.mean()

# Charts

### Chart showing the average course of the car.

```python
bmw_grouped_offers.course.mean().plot(figsize=settings.size)
```

### Chart showing the average price per car year.

```python
bmw_grouped_offers.price.mean().plot(figsize=settings.size)
```

### Pie chart showing percentage share of cars according to the brand. 
Step #1. Create object `grouped_offers_according_to_brand` including grouped offers.

```python
grouped_offers_according_to_brand = offers_df.groupby('brand')
```

Step #2: Show information about group size.

```python
grouped_offers_according_to_brand.size()
```

Step #3: Show pie chart for the object `grouped_offers_according_to_brand.size()`

```python
grouped_offers_according_to_brand.size().plot(kind="pie", label= "")
```

Step #4: Show pie chart for the object `grouped_offers_according_to_brand.size()` with information about percentage share

```python
grouped_offers_according_to_brand.size().plot(kind="pie", autopct='%.2f', label= "")
```
