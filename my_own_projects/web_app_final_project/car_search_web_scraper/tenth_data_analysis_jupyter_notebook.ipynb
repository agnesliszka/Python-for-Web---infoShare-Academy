```python
import pandas as pd
import matplotlib.pyplot as plt
import settings
```

```python
offers_df = pd.read_sql_table('offers','sqlite:///offers.db', index_col='id')
offers_bmw_seria5 = offers_df[offers_df['model'] == 'Seria 5']
bmw_grouped_offers = offers_bmw_seria5.groupby('production_year')

fig, ax = plt.subplots(figsize=settings.size)
ax.set_title("Two axes 'Y', different characteristics")
ax.set_xlabel('Production year')

ax.plot(bmw_grouped_offers.course.mean(), color='blue', marker='+', label='Average course')
ax.set_ylabel('Average course [km]')
ax.tick_params(axis='y', labelcolor='blue')
ax.legend(loc=2)

ax2 = ax.twinx()
ax2.plot(bmw_grouped_offers.price.mean(), color='red', marker='v', label='Average price')
ax2.set_ylabel('Average price [PLN]')
ax2.tick_params(axis='y', labelcolor='red')
ax2.legend(loc=3)
```

### Price comparison of different models

```python
offers_bmw_seria5 = offers_df[offers_df['model'] == 'Seria 5']
bmw_grouped_offers = offers_bmw_seria5.groupby('production_year')
print(bmw_grouped_offers)

offers_opel_corsa = offers_df[offers_df['model'] == 'Corsa']
opel_grouped_offers = offers_opel_corsa.groupby('production_year')

offers_mercedes_klasa_c = offers_df[offers_df['model'] == 'Klasa C']
mercedes_grouped_offers = offers_mercedes_klasa_c.groupby('production_year')

fig, ax = plt.subplots(figsize=settings.size)
ax.set_title("One axe 'Y', comparison of two different brands")
ax.set_xlabel('Production year')
ax.set_ylabel('Average price [PLN]')

ax.plot(bmw_grouped_offers.price.mean(), color='red', marker='+', label='BMW')
ax.plot(opel_grouped_offers.price.mean(), color='blue', marker='o', label='Opel')
ax.plot(mercedes_grouped_offers.price.mean(), color='black', marker='v', label='Mercedes')
ax.legend(loc=2)
```
