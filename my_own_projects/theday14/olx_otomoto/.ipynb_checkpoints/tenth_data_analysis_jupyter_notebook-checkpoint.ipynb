# Wizualizacja danych ofertowych

### Dwie osie 'Y', różne charakterystyki

Przygotuj wykres oparty o tabelę `offers` z bazy sqlite3:
1. wykres zawiera dwie krzywe (średnia cena i średni przebieg) w dziedzinie roku produkcji (oś X) dla wybranego modelu samochodu,
1. każda krzywa ma **swoją** oś Y,
1. uzupełnij wykres odpowiednimi opisami, użyj różnych markerów i różnych kolorów,
1. użyj `group_by` dla kolumny production_year

```python
import pandas as pd
import matplotlib.pyplot as plt
import settings
```

```python
offers_df = pd.read_sql_table('offers','sqlite:///offers.db', index_col='id')
offers_bmw_seria5 = offers_df[offers_df['model'] == 'Seria 5']
bmw_grouped_offers = offers_bmw_seria5.groupby('production_year')

fig, ax = plt.subplots(figsize=settings.size)
ax.set_title("Two axes 'Y', different characteristics")
ax.set_xlabel('Production year')

ax.plot(bmw_grouped_offers.course.mean(), color='blue', marker='+', label='Average course')
ax.set_ylabel('Average course [km]')
ax.tick_params(axis='y', labelcolor='blue')
ax.legend(loc=2)

ax2 = ax.twinx()
ax2.plot(bmw_grouped_offers.price.mean(), color='red', marker='v', label='Average price')
ax2.set_ylabel('Average price [PLN]')
ax2.tick_params(axis='y', labelcolor='red')
ax2.legend(loc=3)
```

### Porównanie cen różnych modeli

Przygotuj wykres oparty o tabelę `offers` z bazy sqlite3:
1. wykres zawiera trzy krzywe (średnia cena) w dziedzinie roku produkcji (oś X) dla trzech wybranych modeli samochodów,
1. wykres ma **jedną** oś Y,
1. uzupełnij wykres odpowiednimi opisami, użyj różnych markerów i różnych kolorów,
1. użyj `group_by` dla kolumny production_year

```python
offers_bmw_seria5 = offers_df[offers_df['model'] == 'Seria 5']
bmw_grouped_offers = offers_bmw_seria5.groupby('production_year')
print(bmw_grouped_offers)

offers_opel_corsa = offers_df[offers_df['model'] == 'Corsa']
opel_grouped_offers = offers_opel_corsa.groupby('production_year')

offers_mercedes_klasa_c = offers_df[offers_df['model'] == 'Klasa C']
mercedes_grouped_offers = offers_mercedes_klasa_c.groupby('production_year')

fig, ax = plt.subplots(figsize=settings.size)
ax.set_title("One axe 'Y', comparison of two different brands")
ax.set_xlabel('Production year')
ax.set_ylabel('Average price [PLN]')

ax.plot(bmw_grouped_offers.price.mean(), color='red', marker='+', label='BMW')
ax.plot(opel_grouped_offers.price.mean(), color='blue', marker='o', label='Opel')
ax.plot(mercedes_grouped_offers.price.mean(), color='black', marker='v', label='Mercedes')
ax.legend(loc=2)
```

### Wykorzystanie `subplot` i zapis wyniku

1. Korzystając z wypracowanych właśnie "przepisów" na wykresy, umieść oba wykresy obok siebie na jednym obszarze (`figure`) 
1. Zapisz obszar z oboma wykresami do pliku PDF. Użyj `fig.savefig('complete_image.pdf')`

```python
fig = plt.figure(figsize=settings.size)

ax11 = fig.add_subplot(1,2,1)
ax21 = fig.add_subplot(1,2,2)

offers_df = pd.read_sql_table('offers','sqlite:///offers.db', index_col='id')
offers_bmw_seria5 = offers_df[offers_df['model'] == 'Seria 5']
bmw_grouped_offers = offers_bmw_seria5.groupby('production_year')

ax11.set_title("Two axes 'Y', different characteristics")
ax11.set_xlabel('Production year')

ax11.plot(bmw_grouped_offers.course.mean(), color='blue', marker='+', label='Average course')
ax11.set_ylabel('Average course [km]')
ax11.tick_params(axis='y', labelcolor='blue')
ax11.legend(loc=2)


ax12.plot(bmw_grouped_offers.price.mean(), color='red', marker='v', label='Average price')
ax12.set_ylabel('Average price [PLN]')
ax12.tick_params(axis='y', labelcolor='red')
ax12.legend(loc=3)

offers_bmw_seria5 = offers_df[offers_df['model'] == 'Seria 5']
bmw_grouped_offers = offers_bmw_seria5.groupby('production_year')
print(bmw_grouped_offers)

offers_opel_corsa = offers_df[offers_df['model'] == 'Corsa']
opel_grouped_offers = offers_opel_corsa.groupby('production_year')

offers_mercedes_klasa_c = offers_df[offers_df['model'] == 'Klasa C']
mercedes_grouped_offers = offers_mercedes_klasa_c.groupby('production_year')

ax21.set_title("One axe 'Y', comparison of two different brands")
ax21.set_xlabel('Production year')
ax21.set_ylabel('Average price [PLN]')

ax21.plot(bmw_grouped_offers.price.mean(), color='red', marker='+', label='BMW')
ax21.plot(opel_grouped_offers.price.mean(), color='blue', marker='o', label='Opel')
ax21.plot(mercedes_grouped_offers.price.mean(), color='black', marker='v', label='Mercedes')
ax21.legend(loc=2)

fig.savefig('complete_image.pdf')
```
